```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

#  wav 
y, sr = librosa.load('bark_generation.wav')

# 
# 
D = librosa.stft(y)

# 
S = librosa.feature.melspectrogram(S=D, sr=sr, n_mels=128)

# 
log_S = librosa.power_to_db(S, ref=np.max)

# 
plt.figure(figsize=(12, 4))

# 
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')

# 
plt.colorbar(format='%+02.0f dB')

# 
plt.title('Mel spectrogram')
plt.tight_layout()

# 
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

#  wav 
y, sr = librosa.load('bark_generation.wav')

# 
D = librosa.stft(y)

# 
S = librosa.feature.melspectrogram(S=D, sr=sr, n_mels=128)

# 
log_S = librosa.power_to_db(S, ref=np.max)

# 
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=13)

# 
fig, ax = plt.subplots(2, 1, figsize=(12, 8))

# 
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel', ax=ax[0])
ax[0].set(title='Mel spectrogram')

# 
librosa.display.specshow(mfcc, sr=sr, x_axis='time', ax=ax[1])
ax[1].set(title='MFCC')

# 
plt.tight_layout()
plt.show()
```
