```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
import os
import torch

# Set the current default CUDA device to GPU 1
torch.cuda.set_device(1)

# Verify the change
default_device = torch.cuda.current_device()
print(f"The current default CUDA device is: {default_device}")
# Specify the GPU ID to use
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
# download and load all models
preload_models()

# generate audio from text
text_prompt = """
     "Eat it a little more," he said. "Eat it well."
"""
audio_array = generate_audio(text_prompt,history_prompt="v2/en_speaker_9")

# save audio to disk
write_wav("bark_generation.wav", SAMPLE_RATE, audio_array)
  
# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
zh_fix_path="/home/ydoit/AIGC/Raw-ZH/zh_normal_fix.txt"
zh_random_path="/home/ydoit/AIGC/Raw-ZH/zh_normal_random.txt"
en_fix_path="/home/ydoit/AIGC/Raw-EN/en_normal_fix.txt"
en_random_path="/home/ydoit/AIGC/Raw-EN/en_normal_random.txt"
```

```python
with open(zh_fix_path, "r") as f:
    zh_fix_text = f.readlines()
with open(zh_random_path, "r") as f:
    zh_random_text = f.readlines()[100:200]
with open(en_fix_path, "r") as f:
    en_fix_text = f.readlines()
with open(en_random_path, "r") as f:
    en_random_text = f.readlines()[200:300]
```

```python
# zh_fix_path="/home/ydoit/AIGC/Raw-ZH/zh_normal_fix.txt"
# en_text_path="en_text.txt"
for i in range(100):
    try :
        sp_index=int(i/10)
        text_index=int(i%10)+1
        zh_fix=zh_fix_text[i]
        zh_random=zh_random_text[i]
        en_fix=en_fix_text[i]
        en_random=en_random_text[i]

        audio_array = generate_audio(zh_fix,history_prompt=f"v2/zh_speaker_{sp_index}")
        write_wav(f"output/fix/zh/speaker{sp_index+1}/audio{text_index}.wav", SAMPLE_RATE, audio_array)
        # print(f"zh_{count}.wav")
        with open(f"output/fix/zh/speaker{sp_index+1}/audio{text_index}.txt","w") as f:
            f.write(zh_fix)
        
        
        audio_array = generate_audio(en_fix,history_prompt=f"v2/en_speaker_{sp_index}")
        write_wav(f"output/fix/en/speaker{sp_index+1}/audio{text_index}.wav", SAMPLE_RATE, audio_array)
        # print(f"zh_{count}.wav")
        with open(f"output/fix/en/speaker{sp_index+1}/audio{text_index}.txt","w") as f:
            f.write(en_fix)
            
        audio_array = generate_audio(zh_random,history_prompt=f"v2/zh_speaker_{sp_index}")
        write_wav(f"output/random/zh/speaker{sp_index+1}/audio{text_index}.wav", SAMPLE_RATE, audio_array)
        # print(f"zh_{count}.wav")
        with open(f"output/random/zh/speaker{sp_index+1}/audio{text_index}.txt","w") as f:
            f.write(zh_random)
        
        
        audio_array = generate_audio(en_random,history_prompt=f"v2/en_speaker_{sp_index}")
        write_wav(f"output/random/en/speaker{sp_index+1}/audio{text_index}.wav", SAMPLE_RATE, audio_array)
        # print(f"zh_{count}.wav")
        with open(f"output/random/en/speaker{sp_index+1}/audio{text_index}.txt","w") as f:
            f.write(en_random)
        print(f"finish {i}")
    except Exception as e:
        print(e)
        print("stop at :"+str(i))
exit()
```
