```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
import os
import torch

# Set the current default CUDA device to GPU 1
torch.cuda.set_device(1)

# Verify the change
default_device = torch.cuda.current_device()
print(f"The current default CUDA device is: {default_device}")
# Specify the GPU ID to use
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
# download and load all models
preload_models()

# generate audio from text
text_prompt = """
     "Eat it a little more," he said. "Eat it well."
"""
audio_array = generate_audio(text_prompt,history_prompt="v2/en_speaker_9")

# save audio to disk
write_wav("bark_generation.wav", SAMPLE_RATE, audio_array)
  
# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
zh_text_path="zh_text.txt"
en_text_path="en_text.txt"
with open(zh_text_path,"r") as f:
    zh_text=f.readlines()
    count=0
    for text in zh_text:
        sp=int(count/10)
        index=int(count%10)
        audio_array = generate_audio(text,history_prompt=f"v2/zh_speaker_{sp}")
        write_wav(f"output/zh/speaker{sp+1}/audio{index+1}.wav", SAMPLE_RATE, audio_array)
        # print(f"zh_{count}.wav")
        with open(f"output/zh/speaker{sp+1}/audio{index+1}.txt","w") as f:
            f.write(text)
        count+=1
with open(en_text_path,"r") as f:
    en_text=f.readlines()
    count=0
    for text in en_text:
        sp=int(count/10)
        index=int(count%10)
        audio_array = generate_audio(text,history_prompt=f"v2/en_speaker_{sp}")
        write_wav(f"output/en/speaker{sp+1}/audio{index+1}.wav", SAMPLE_RATE, audio_array)
        # print(f"en_{count}.wav")
        with open(f"output/en/speaker{sp+1}/audio{index+1}.txt","w") as f:
            f.write(text)
        count+=1
```
